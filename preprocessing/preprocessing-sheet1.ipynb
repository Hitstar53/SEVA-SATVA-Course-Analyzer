Concatenating all excel sheets in a single CSV file (for even sem 23-24):

```python
import pandas as pd
import os
data_dir = r'..\data'
```

Courses:

```python
courses = pd.read_csv(os.path.join(data_dir, 'courses.csv'))
courses
```

```python
course_names = courses['course_name']
```

Loading odd sem 22-23.xlsx:

```python
df = pd.read_excel(os.path.join(data_dir, 'odd sem 2022-23.xlsx'), sheet_name=1)
```

```python
df.head(10)
```

```python
final = pd.DataFrame(columns=['email', 'name', 'uid', 'branch', 'sem', 'course'])
for sheet_no in range(21):
    df = pd.read_excel(os.path.join(data_dir, 'odd sem 2022-23.xlsx'), sheet_name=sheet_no)
    df.columns = ['email', 'name', 'uid', 'class', 'branch', 'course', 'sem']
    df.drop(columns=['class'], inplace=True)
    df = df[['email', 'name', 'uid', 'branch', 'sem', 'course']]
    final = pd.concat([final, df], ignore_index=True)
```

```python
final.isnull().sum()
```

```python
final.shape
```

```python
final.to_csv(os.path.join(data_dir, 'sheet-1.csv'), index=False)
```

```python
courses = pd.read_csv(os.path.join(data_dir, 'courses.csv'))
course_names = courses['course_name'].tolist()
course_names
```

After cleaning some data in excel, normalizing rest of the courses:

```python
sheet1 = pd.read_csv(os.path.join(data_dir, 'sheet-1.csv'))
sheet1['course'].unique()
```

```python
not_matching = [i for i in sheet1['course'].unique() if i not in course_names]
not_matching
```

Deleting the rows with 'Course Allocated' in course column:

```python
sheet1.drop(sheet1[sheet1['course'] == 'Course Allocated'].index, inplace=True)
not_matching = [i for i in sheet1['course'].unique() if i not in course_names]
not_matching
```

Converting all names to title case:

```python
sheet1['name'] = sheet1['name'].map(lambda x: x.strip().title())
```

```python
sheet1.head(20)
```

```python
sheet1.sort_values(by=['course', 'sem', 'branch', 'uid'], ascending=True, inplace=True)
sheet1.head(20)
```

```python
sheet1.to_csv(os.path.join(data_dir, 'sheet-1.csv'), index=False)
```

Deleting duplicate rows:

```python
print(f'Number of duplicate rows: {sheet1.duplicated().sum()} / {sheet1.shape[0]}')
```

```python
sheet1.loc[sheet1.duplicated(), :]
```

```python
sheet1.drop_duplicates(inplace=True, keep='first', ignore_index=True)
```

```python
sheet1.shape[0]
```

```python
sheet1.to_csv(os.path.join(data_dir, 'sheet-1.csv'), index=False)
```

Checking for duplicates based on email:

```python
sheet1.duplicated(subset=['email']).sum()
```

```python
duplicate_emails = set(sheet1['email'][sheet1.duplicated(subset=['email'])].values)
duplicated_rows = sheet1[sheet1['email'].map(lambda x: x in duplicate_emails)].sort_values(by='email')
duplicated_rows
```

```python
duplicated_rows.to_csv(os.path.join(data_dir, 'sheet-1-duplicated-rows.csv'), index=False)
```

Joiining with the courses df:

```python
cols_to_keep = list(sheet1.columns) + ['category']
sheet1 = pd.merge(sheet1, courses, left_on='course', right_on='course_name')[cols_to_keep]
sheet1
```

```python
sheet1.to_csv(os.path.join(data_dir, 'sheet-1.csv'), index=False)
```

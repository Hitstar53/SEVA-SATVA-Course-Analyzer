```python
final = pd.DataFrame(columns=['email', 'name', 'uid', 'branch', 'sem', 'course'])
for sheet_no in range(21): 
    df = pd.read_excel(os.path.join(data_dir, 'even sem 2022-23.xlsx'), sheet_name=sheet_no, usecols=['Email Address', 'Name (as in the college records)', 'UID No.', 'Class (to be entered in next Semester)', 'Branch', 'SEVA/SATVA Course Name', 'Sem'])
    df.columns = ['email', 'name', 'uid', 'class', 'branch', 'course', 'sem']
    df.drop(columns=['class'], inplace=True)
    df = df[['email', 'name', 'uid', 'branch', 'sem', 'course']]
    final = pd.concat([final, df], ignore_index=True)
```

```python
courses = pd.read_csv(os.path.join(data_dir, 'courses.csv'))
course_names = courses['course_name'].tolist()
course_names
```

After cleaning some data in excel, normalizing rest of the courses:

```python
sheet2 = pd.read_csv(os.path.join(data_dir, 'sheet-2.csv'))
sheet2['course'].unique()
```

```python
not_matching = [i for i in sheet2['course'].unique() if i not in course_names]
not_matching
```

```python
course_names
```

Stripping leading and trailing spaces:

```python
sheet2['course'] = sheet2['course'].str.strip()
not_matching = [i for i in sheet2['course'].unique() if i not in course_names]
not_matching
```

NSS related issues:

```python
sheet2['course'] = sheet2['course'].map(lambda x: 'NSS' if 'nss' in x.lower() else x)
not_matching = [i for i in sheet2['course'].unique() if i not in course_names]
not_matching
```

Keyboard playing:

```python
sheet2['course'] = sheet2['course'].map(lambda x: 'Basics of Keyboard playing' if 'keyboard' in x.lower() else x)
not_matching = [i for i in sheet2['course'].unique() if i not in course_names]
not_matching
```

Idea lab:

```python
sheet2['course'] = sheet2['course'].map(lambda x: 'Idea Lab' if x.startswith('Idea') else x)
not_matching = [i for i in sheet2['course'].unique() if i not in course_names]
not_matching
```

Film appreciation:

```python
sheet2['course'] = sheet2['course'].map(lambda x: 'Film Appreciation' if 'appreciation' in x.lower() else x)
not_matching = [i for i in sheet2['course'].unique() if i not in course_names]
not_matching
```

```python
def normalize(course_name):
    if course_name == 'Study of book':
        return 'Read a book'
    if course_name == 'Working in NGO':
        return 'Working with NGO'
    if course_name == 'Movie review':
        return 'Film Appreciation'
    if course_name == 'Trek':
        return 'Trekking'
    if course_name == 'mentorship to juniors':
        return 'Mentorship to Juniors'
    if course_name == 'Innovation and creativity':
        return 'Innovation and Creativity'
    return course_name

sheet2['course'] = sheet2['course'].map(normalize)
not_matching = [i for i in sheet2['course'].unique() if i not in course_names]
not_matching
```

Converting all names to title case:

```python
sheet2['name'] = sheet2['name'].map(lambda x: x.strip().title())
```

```python
sheet2.head(20)
```

```python
sheet2.sort_values(by=['course', 'sem', 'branch', 'uid'], ascending=True, inplace=True)
sheet2.head(20)
```

```python
sheet2.to_csv(os.path.join(data_dir, 'sheet-2.csv'), index=False)
```

Checking for duplicates based on email:

```python
duplicate_emails = set(sheet2['email'][sheet2.duplicated(subset=['email'])].values)
duplicated_rows = sheet2[sheet2['email'].map(lambda x: x in duplicate_emails)].sort_values(by='email')
duplicated_rows
```

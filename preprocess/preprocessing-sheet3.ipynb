Concatenating all excel sheets in a single CSV file (for even sem 23-24):

```python
import pandas as pd
import os
data_dir = '..\data'
```

Courses:

```python
courses = pd.read_csv(os.path.join(data_dir, 'courses.csv'))
courses
```

```python
course_names = courses['course_name']
```

Loading odd sem 23-24.xlsx:

```python
df = pd.read_excel(os.path.join(data_dir, 'odd sem 2023-24.xlsx'), sheet_name=1)
```

```python
df.head(10)
```

```python
final = pd.DataFrame(columns=['email', 'name', 'uid', 'branch', 'sem', 'course'])
for sheet_no in range(1, 27): # skipping instructions sheet
    df = pd.read_excel(os.path.join(data_dir, 'odd sem 2023-24.xlsx'), sheet_name=sheet_no)
    df.columns = ['email', 'name', 'uid', 'class', 'branch', 'course', 'sem']
    df.drop(columns=['class'], inplace=True)
    df = df[['email', 'name', 'uid', 'branch', 'sem', 'course']]
    final = pd.concat([final, df], ignore_index=True)
```

```python
final.isnull().sum()
```

```python
final.shape
```

```python
na_rows = list(set(final.index) - set(final.dropna().index))
final.loc[na_rows, :]
```

Dropping missing values:

```python
final.loc[na_rows, :].shape[0]
```

```python
final.dropna(inplace=True)
final.isnull().sum()
```

```python
final.to_csv(os.path.join(data_dir, 'sheet-3.csv'), index=False)
```

```python
courses = pd.read_csv(os.path.join(data_dir, 'courses.csv'))
course_names = courses['course_name'].tolist()
course_names
```

After cleaning some data in excel, normalizing rest of the courses:

```python
sheet3 = pd.read_csv(os.path.join(data_dir, 'sheet-3.csv'))
sheet3['course'].unique()
```

```python
not_matching = [i for i in sheet3['course'].unique() if i not in course_names]
not_matching
```

These names have to be the same as that in courses.csv

```python
course_names
```

Correcting trekking related spelling mistakes:

```python
sheet3['course'] = sheet3['course'].map(lambda x: 'Trekking' if 'trek' in x.lower() else x)
not_matching = [i for i in sheet3['course'].unique() if i not in course_names]
not_matching
```

Changing all study of life courses to 'Study of Life':

```python
sheet3['course'] = sheet3['course'].map(lambda x: 'Study of Life' if 'study of life' in x.lower() else x)
not_matching = [i for i in sheet3['course'].unique() if i not in course_names]
not_matching
```

Trimming all leading and trailing spaces:

```python
sheet3['course'] = sheet3['course'].str.strip()
not_matching = [i for i in sheet3['course'].unique() if i not in course_names]
not_matching
```

Normalizing all sports related courses:

```python
sheet3['course'] = sheet3['course'].map(lambda x: 'Sports' if 'sports' in x.lower() else x)
not_matching = [i for i in sheet3['course'].unique() if i not in course_names]
not_matching
```

Correcting 'Intro to bhagvadgita':

```python
sheet3['course'] = sheet3['course'].map(lambda x: 'Introduction to Bhagvadgita' if x.lower().startswith('intro') else x)
not_matching = [i for i in sheet3['course'].unique() if i not in course_names]
not_matching
```

Removing 'online':

```python
sheet3['course'] = sheet3['course'].map(lambda x: 'Introduction to Bhagvadgita' if '(online' in x.lower() else x)
not_matching = [i for i in sheet3['course'].unique() if i not in course_names]
not_matching
```

Running:

```python
sheet3['course'] = sheet3['course'].map(lambda x: 'Running' if x.lower().startswith('run') else x)
not_matching = [i for i in sheet3['course'].unique() if i not in course_names]
not_matching
```

```python
def normalize(course_name):
    if course_name == 'Innovative and Creativity':
        return 'Innovation and Creativity'
    if course_name == 'Basic of Fire Safety':
        return 'Basics of Fire Safety'
    if course_name == 'Flim appreciation':
        return 'Film Appreciation'
    if course_name == 'Fundamentals of Photography.':
        return 'Fundamentals of Photography'
    return course_name

sheet3['course'] = sheet3['course'].map(normalize)
not_matching = [i for i in sheet3['course'].unique() if i not in course_names]
not_matching
```

Converting all names to title case:

```python
sheet3['name'] = sheet3['name'].str.title()
```

```python
sheet3.head(20)
```

```python
sheet3.sort_values(by=['course', 'sem', 'branch', 'uid'], ascending=True, inplace=True)
sheet3.head(20)
```

```python
sheet3.to_csv(os.path.join(data_dir, 'sheet-3.csv'), index=False)
```

```python
final = pd.DataFrame(columns=['email', 'name', 'uid', 'branch', 'sem', 'course'])
for sheet_no in range(1, 27): # skipping instructions sheet
    df = pd.read_excel(os.path.join(data_dir, 'odd sem 2023-24.xlsx'), sheet_name=sheet_no)
    df.columns = ['email', 'name', 'uid', 'class', 'branch', 'course', 'sem']
    df.drop(columns=['class'], inplace=True)
    df = df[['email', 'name', 'uid', 'branch', 'sem', 'course']]
    final = pd.concat([final, df], ignore_index=True)
```

```python
sheet4['course'] = sheet4['course'].map(lambda x: 'Introduction to Bhagvadgita' if x.lower().startswith('introduction') else x)
not_matching = [i for i in sheet4['course'].unique() if i not in course_names]
not_matching
```

```python
sheet4['course'] = sheet4['course'].str[0].map(lambda x: x.upper()).values + sheet4['course'].str[1:]
not_matching = [i for i in sheet4['course'].unique() if i not in course_names]
not_matching
```

```python
sheet4['course'] = sheet4['course'].str.strip()
not_matching = [i for i in sheet4['course'].unique() if i not in course_names]
not_matching
```

```python
sheet4['course'] = sheet4['course'].map(lambda x: 'Sports' if x.lower().startswith('sports') else x)
not_matching = [i for i in sheet4['course'].unique() if i not in course_names]
not_matching
```

```python
def normalize(course_name):
    if course_name == 'Design thinking':
        return 'Design Thinking'
    if course_name == 'Skipiing':
        return 'Skipping'
    if course_name == 'Staircase Climbing.':
        return 'Staircase Climbing'
    if course_name == 'Bhagwan Gita':
        return 'Introduction to Bhagvadgita'
    return course_name

sheet4['course'] = sheet4['course'].map(normalize)
not_matching = [i for i in sheet4['course'].unique() if i not in course_names]
not_matching
```

```python
sheet4['name'] = sheet4['name'].str.title()
```
